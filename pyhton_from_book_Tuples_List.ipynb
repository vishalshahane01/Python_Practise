```python
num_cats = 100
cats_with_hats = [False] * num_cats

# Round 1: Stop at every cat, placing a hat on each one.
for i in range(num_cats):
  cats_with_hats[i] = True

print("After Round 1:")
print(f"Cats with hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if has_hat]}")
print(f"Cats without hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if not has_hat]}")
```

```python
# Round 2: Stop at every second cat, toggling its hat status.
# Cats are 1-indexed in the problem description, so we iterate starting from index 1 (cat #2) and step by 2.
for i in range(1, num_cats, 2):
  cats_with_hats[i] = not cats_with_hats[i]

print("After Round 2:")
print(f"Cats with hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if has_hat]}")
print(f"Cats without hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if not has_hat]}")
```

```python
# Round 3: Stop at every third cat, toggling its hat status.
# Cats are 1-indexed in the problem description, so we iterate starting from index 2 (cat #3) and step by 3.
for i in range(2, num_cats, 3):
  cats_with_hats[i] = not cats_with_hats[i]

print("\nAfter Round 3:")
print(f"Cats with hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if has_hat]}")
print(f"Cats without hats: {[i + 1 for i, has_hat in enumerate(cats_with_hats) if not has_hat]}")
```

```python
num_cats = 100
cats_with_hats = [False] * num_cats

# Simulate all 100 rounds
for round_num in range(1, num_cats + 1):
  # For each round, stop at every `round_num`-th cat
  # Cats are 1-indexed in the problem description, so list index is `round_num - 1`, `2*round_num - 1`, etc.
  for i in range(round_num - 1, num_cats, round_num):
    cats_with_hats[i] = not cats_with_hats[i]

  # Print status after each round
  print(f"\n--- After Round {round_num} ---")
  cats_with = [i + 1 for i, has_hat in enumerate(cats_with_hats) if has_hat]
  cats_without = [i + 1 for i, has_hat in enumerate(cats_with_hats) if not has_hat]
  print(f"Cats with hats: {cats_with}")
  print(f"Cats without hats: {cats_without}")

# Get the final list of cats with hats (1-indexed for readability)
final_cats_with_hats = [i + 1 for i, has_hat in enumerate(cats_with_hats) if has_hat]

print(f"\n=== Final Result After all {num_cats} rounds ===")
print(f"Cats with hats: {final_cats_with_hats}")
print(f"Total number of cats with hats: {len(final_cats_with_hats)}")
```
